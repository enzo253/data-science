```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))

x = [1,2,3,4]
y = [4,5,6,7]

plt.title('not our first graph', fontdict={'fontname' : 'Times New Roman', 'fontsize' : 15})
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')

plt.yticks([1,2,3,4,5,6])
#plt.xticks([1,3,5,7,8])

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2, color = 'red', label = 'x2')
plt.plot(x, y, label = '2x', color = 'red', linewidth = 1.5, marker = '.', markersize = 5, linestyle = '--' )
plt.legend()
plt.show()
```
